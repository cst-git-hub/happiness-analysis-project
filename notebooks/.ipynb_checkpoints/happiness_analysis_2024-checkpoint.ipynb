# Global Happiness Analysis


Research Question

Do institutional quality and life conditions
explain cross-country differences in happiness?


This project explores the relationship between:

- corruption perception (2024)
- democracy index (2024)
- life expectancy (2026)

and national happiness levels (2024).


Used abbreviations for indicators:

- CPI = corruption perceptions index
- DI = democracy index
- LE (ys) = life expectancy (years)
- SrLS = self-reported life satisfaction (happiness)


```python
import plotly.io as pio
pio.renderers.default = "notebook_connected"


# Load cleaned datasets

import pandas as pd
from pathlib import Path
import plotly.express as px

BASE_DIR = Path.cwd().parents[0]

data_dir = BASE_DIR / "data" / "clean"

corruption = pd.read_csv(data_dir / "corruption_cleaned.csv")
democracy = pd.read_csv(data_dir / "democracy_cleaned.csv")
life_exp = pd.read_csv(data_dir / "life_expectancy_cleaned.csv")
happiness = pd.read_csv(data_dir / "happiness_cleaned.csv")


# Merge datasets
# rename and drop unnecessary country column
# highlight important countries

highlight = [
    "USA","GBR","JPN","FIN","DEU", "FRA",
    "CHN","IND","RUS","CHE","HUN","AUT"
]

corr_df = pd.merge(corruption, happiness, on="icc", how="inner")
corr_df = corr_df.rename(columns={"country_x": "country"})
corr_df = corr_df.drop(columns=["country_y"])
corr_df["label"] = corr_df["icc"].apply(
    lambda x: f"<b>{x}</b>" if x in highlight else ""
) # country labels with bold letters

demo_df = pd.merge(democracy, happiness, on="icc", how="inner")
demo_df = demo_df.rename(columns={"country_x": "country"})
demo_df = demo_df.drop(columns=["country_y"])
demo_df["label"] = demo_df["icc"].apply(
    lambda x: f"<b>{x}</b>" if x in highlight else "")

life_df = pd.merge(life_exp, happiness, on="icc", how="inner")
life_df = life_df.rename(columns={"country_x": "country"})
life_df = life_df.drop(columns=["country_y"])
life_df["label"] = life_df["icc"].apply(
    lambda x: f"<b>{x}</b>" if x in highlight else "")

    
## Correlation calcs

corr_corr = corr_df[["CPI", "SrLS"]].corr().iloc[0,1]
demo_corr = demo_df[["DI", "SrLS"]].corr().iloc[0,1]
life_corr = life_df[["LE (ys)", "SrLS"]].corr().iloc[0,1]

print("---------------------------------------------", "\n")
print("Results of correlation calculation in three relations:", "\n")

print(f"Corruption vs Happiness correlation: {corr_corr:.2f}")
print(f"Democracy vs Happiness correlation: {demo_corr:.2f}")
print(f"Life Expectancy vs Happiness correlation: {life_corr:.2f}", "\n")
print("---------------------------------------------")


## Correlation heatmap

merged = corr_df[["icc", "SrLS", "CPI"]] \
    .merge(demo_df[["icc", "DI"]], on="icc") \
    .merge(life_df[["icc", "LE (ys)"]], on="icc")

corr = merged[["SrLS", "CPI", "DI", "LE (ys)"]].corr()   # Correlation matrix

import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

mask = np.eye(len(corr), dtype=bool)

plt.figure()

sns.heatmap(
    corr,
    annot=True,
    cmap="Blues",
    fmt=".2f",
    vmin=0,
    vmax=1,
    mask=mask,     # hide diagonal
    linewidths=0.5
)

plt.title("Correlation Matrix Heatmap")
plt.show()

print("---------------------------------------------")


# Interactive scatter plots incl. trendline

fig1 = px.scatter(
    corr_df,
    x="CPI",
    y="SrLS",
    hover_name="country",
    text="label",
    trendline="ols",
    color_discrete_sequence=["darkred"],
    trendline_color_override="royalblue",
#    title=f"Corruption vs Happiness (corr={corr_corr:.2f})"
)

results = px.get_trendline_results(fig1)
model = results.iloc[0]["px_fit_results"]

slope = model.params[1]
intercept = model.params[0]
r2 = model.rsquared
p = model.pvalues[1]

fig1.update_layout(
    title=f"Corruption vs Happiness | y={slope:.2f}x+{intercept:.2f} | R²={r2:.2f} | p={p:.4f}",
    title_x=0.5
)


fig1.show()


fig2 = px.scatter(
    life_df,
    x="LE (ys)",
    y="SrLS",
    hover_name="country",
    text="label",
    trendline="ols",
    color_discrete_sequence=["royalblue"],
    trendline_color_override="darkred",
#    title=f"Life Expectancy vs Happiness (corr={life_corr:.2f})"
)

results = px.get_trendline_results(fig2)
model = results.iloc[0]["px_fit_results"]

slope = model.params[1]
intercept = model.params[0]
r2 = model.rsquared
p = model.pvalues[1]

fig2.update_layout(
    title=f"Life Expectancy vs Happiness | y={slope:.2f}x+{intercept:.2f} | R²={r2:.2f} | p={p:.4f}",
    title_x=0.5
)

fig2.show()



fig3 = px.scatter(
    demo_df,
    x="DI",
    y="SrLS",
    hover_name="country",
    text="label",
    trendline="ols",
    color_discrete_sequence=["green"],
    trendline_color_override="black",
#    title=f"Democracy vs Happiness (corr={demo_corr:.2f})"
)

results = px.get_trendline_results(fig3)
model = results.iloc[0]["px_fit_results"]

slope = model.params[1]
intercept = model.params[0]
r2 = model.rsquared
p = model.pvalues[1]

fig3.update_layout(
    title=f"Democracy vs Happiness | y={slope:.2f}x+{intercept:.2f} | R²={r2:.2f} | p={p:.4f}",
    title_x=0.5
)

fig3.show()


print("-----------------3 graphs dahsboard with same y axis (Happiness)------------------------")
#import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go

fig = make_subplots(
    rows=1,
    cols=3,
    subplot_titles=[
        "Corruption vs Happiness",
        "Democracy vs Happiness",
        "Life Expectancy vs Happiness"
    ]
)

fig1 = px.scatter(
    corr_df,
    x="CPI",
    y="SrLS",
    hover_name="country",
    trendline="ols",
    color_discrete_sequence=["darkred"],
    trendline_color_override="royalblue"
)

for trace in fig1.data:
    fig.add_trace(trace, row=1, col=1)

fig2 = px.scatter(
    demo_df,
    x="DI",
    y="SrLS",
    hover_name="country",
    trendline="ols",
    color_discrete_sequence=["green"],
    trendline_color_override="black"
)

for trace in fig2.data:
    fig.add_trace(trace, row=1, col=2)

fig3 = px.scatter(
    life_df,
    x="LE (ys)",
    y="SrLS",
    hover_name="country",
    trendline="ols",
    color_discrete_sequence=["royalblue"],
    trendline_color_override="darkred"
)

for trace in fig3.data:
    fig.add_trace(trace, row=1, col=3)

fig.update_layout(
    template="plotly_white",
    title="Factors (X) Associated with National Happiness (Y)",
    title_x=0.5,
    height=500
)

fig.show()


print("-----------------Interactive World Map Charts (choropleth map charts)------------------------")

#import plotly.express as px

fig1 = px.choropleth(
    happiness,
    locations="icc",          # ISO3 country code
    color="SrLS",             # happiness index
    hover_name="country",
    color_continuous_scale="Brwnyl",
    title="Global Happiness Distribution, 2024"
)
fig1.update_layout(
    geo=dict(
        showframe=False,
        showcoastlines=True,
        projection_type="natural earth"
    )
)
fig1.show()


fig2 = px.choropleth(
    corruption,
    locations="icc",
    color="CPI",
    hover_name="country",
    color_continuous_scale="Reds",
    title="Corruption Perception Index, 2024"
)
fig2.update_layout(
    geo=dict(
        showframe=False,
        showcoastlines=True,
        projection_type="natural earth"
    )
)
fig2.show()


fig3 = px.choropleth(
    democracy,
    locations="icc",
    color="DI",
    hover_name="country",
    color_continuous_scale="Blues",
    title="Democracy Index, 2024"
)
fig3.update_layout(
    geo=dict(
        showframe=False,
        showcoastlines=True,
        projection_type="natural earth"
    )
)
fig3.show()


fig4 = px.choropleth(
    life_exp,
    locations="icc",
    color="LE (ys)",
    hover_name="country",
    color_continuous_scale="Greens",
    title="Life Expectancy, 2026"
)
fig4.update_layout(
    geo=dict(
        showframe=False,
        showcoastlines=True,
        projection_type="natural earth"
    )
)
fig4.show()

```
